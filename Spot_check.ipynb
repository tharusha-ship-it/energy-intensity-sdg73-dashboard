```python
import pandas as pd

df = pd.read_csv("energy_intensity_dashboard.csv")

print("Entity type counts:")
print(df["entity_type"].value_counts())

print("\nSample Country labels:")
print(
    df[df["entity_type"] == "Country"]["entity_name"]
    .drop_duplicates()
    .sort_values()
    .head(20)
    .to_list()
)

print("\nSample Region/Aggregate labels:")
print(
    df[df["entity_type"] == "Region/Aggregate"]["entity_name"]
    .drop_duplicates()
    .sort_values()
    .head(20)
    .to_list()
)
```

```python
import pandas as pd

df = pd.read_csv("energy_intensity_dashboard.csv")

suspicious = df[
    (df["entity_type"] == "Country") &
    (
        df["entity_name"].str.contains(
            r"World|income|Asia|Africa|Europe|Caribbean|states|OECD|Union|Pacific|Middle East|North America|Sub-Saharan",
            case=False,
            na=False
        )
    )
][["entity_name", "entity_type"]].drop_duplicates().sort_values("entity_name")

print("Possible misclassified rows:")
print(suspicious)
```
